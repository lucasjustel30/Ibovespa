```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

doc = pd.read_excel('dados_acoes.xlsx', sheet_name='PETR4')
plt.plot(doc["date"], doc["open"], color='red')
plt.title('Stock price chart - PETR4', size=20)
plt.legend(['PETR4'], loc='lower center')
plt.xlabel('2022' ,  size=15)
plt.grid()
```

Stock price BBAS3 - Base 100

```python

doc = pd.read_excel('dados_acoes.xlsx' , sheet_name='BBAS3')

plt.plot(doc['date'] , doc['base100'], color='blue')
plt.legend(['BBAS3'], loc='lower center')
plt.title('Stock price - BBAS3 base100', size=20)
plt.grid()

```

Benchmark = ELET6 x IBOV 

```python

doc = pd.read_excel('dados_acoes.xlsx', sheet_name='ELET6') 
plt.plot(doc['date'],doc['base100'], color='red')

doc = pd.read_excel('dados_acoes.xlsx' , sheet_name='IBOV')
plt.plot(doc['date'] , doc['base100'] , color='blue')


plt.title('Benchmark - ELET6', size=20)
plt.legend(['ELET6','IBOV'], loc='upper left')
plt.grid()
   
```
